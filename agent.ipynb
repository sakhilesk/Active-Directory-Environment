```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
search_tool = SerperDevTool()
```

```python
from langchain_openai import ChatOpenAI
```

```python
# Define your agents with roles and goals
researcher = Agent(
    role = "Senior Research Assistant",
    goal = "Look up the latest advancements in AI",
    backstory = """You work at a leading tech think tank,
    your expertise lies in searching Google for AI Agent frameworks.
    You have a knack for dissecting complex data and presenting actionable insight.""",
    verbose=False,
    allow_delegation=False,
    tools=[search_tool],
    llm=ChatOpenAI(model_name="gpt-4-turbo-preview", temperature=0.2)

)
```

```python
writer = Agent(
  role='Professinal short article writer',
  goal='summarize the latest advancements in Ai agents in a concise article',
  backstory="""You are a renowned content Strategist, known for your insightful and articles
  you transform complex concept into compelling narrative""",
  verbose=False,
  allow_delegation=True,
  llm=ChatOpenAI(model_name="gpt-4-turbo-preview", temperature=0.7)
)
```

```python
# Create tasks for your agents
task1 = Task(
 description="""Conduct a comprehensive analysis of the latest advancements in AI Agents in April
 Identify key trends, breakthrough technologies, and potential industry impactful""",
 expected_output="Full analysis report in bullet points",
 agent=researcher
)
```

```python
task2 = Task(
 description="""Using the insights provided, write a short article
 that highlights the most significant AI Agent advancements.
 Your post should be informative yet accessible, catering to a tech-savy audience
 Make it sound cool, avoid complex words so it doesn't sound like AI""",
 expected_output="Full blog post of at least 3 paragraphs",
 agent=writer
)
```

```python
#Instantiate your crew with a sequential process
crew = Crew(
    agents = [researcher, writer],
    tasks = [task1, task2],
    verbose = 1, # You can set it to 1 or 2 to different logging levels
)
```

```python
result = crew.kickoff()

print("#############")
print(result)
```
